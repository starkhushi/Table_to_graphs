```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_ensemble = {
    "Technique": ["Bagging", "Boosting", "Stacking"],
    "Accuracy (%)": [87.24, 91.95, 94.92],
    "Precision (%)": [85.87, 89.23, 93.76],
    "Recall (%)": [83.65, 88.43, 93.92],
    "F-Measure (%)": [84.55, 88.50, 92.65]
}

df_ensemble = pd.DataFrame(data_ensemble)
```

```python
# 1. Accuracy Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='Technique', y='Accuracy (%)', data=df_ensemble, palette='viridis')
plt.title('Accuracy of Ensemble Classifier Techniques')
plt.xlabel('Technique')
plt.ylabel('Accuracy (%)')
plt.savefig('ensemble_accuracy_plot.png')
plt.show()
```

```python
# 2. Precision, Recall, and F-Measure Comparison
plt.figure(figsize=(10, 6))
df_melted = df_ensemble.melt(id_vars='Technique', value_vars=['Precision (%)', 'Recall (%)', 'F-Measure (%)'],
                             var_name='Metric', value_name='Value (%)')
sns.barplot(x='Technique', y='Value (%)', hue='Metric', data=df_melted, palette='coolwarm')
plt.title('Precision, Recall, and F-Measure Comparison')
plt.xlabel('Technique')
plt.ylabel('Value (%)')
plt.legend(title='Metric')
plt.savefig('ensemble_prf_comparison_plot.png')
plt.show()
```

```python
# 3. Line Plot for All Metrics
plt.figure(figsize=(10, 6))
plt.plot(df_ensemble['Technique'], df_ensemble['Accuracy (%)'], marker='o', label='Accuracy (%)')
plt.plot(df_ensemble['Technique'], df_ensemble['Precision (%)'], marker='o', label='Precision (%)')
plt.plot(df_ensemble['Technique'], df_ensemble['Recall (%)'], marker='o', label='Recall (%)')
plt.plot(df_ensemble['Technique'], df_ensemble['F-Measure (%)'], marker='o', label='F-Measure (%)')
plt.title('Ensemble Classifier Metrics Comparison')
plt.xlabel('Technique')
plt.ylabel('Percentage (%)')
plt.legend()
plt.grid(True)
plt.savefig('ensemble_line_plot.png')
plt.show()
```

```python
# 4. Stacked Bar Plot for Metrics
df_ensemble.set_index('Technique')[['Accuracy (%)', 'Precision (%)', 'Recall (%)', 'F-Measure (%)']].plot(
    kind='bar', stacked=True, figsize=(10, 6), colormap='tab20')
plt.title('Stacked Bar Plot of Ensemble Classifier Metrics')
plt.xlabel('Technique')
plt.ylabel('Percentage (%)')
plt.legend(title='Metric')
plt.savefig('ensemble_stacked_bar_plot.png')
plt.show()
```

```python
# 5. Heatmap of Performance Metrics
plt.figure(figsize=(8, 6))
heatmap_data_ensemble = df_ensemble.set_index('Technique')
sns.heatmap(heatmap_data_ensemble, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Heatmap of Ensemble Classifier Metrics')
plt.savefig('ensemble_heatmap.png')
plt.show()
```
