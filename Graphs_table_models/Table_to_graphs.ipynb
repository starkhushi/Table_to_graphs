```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data from the table
data = {
    "Technique": ["NB", "RF", "LR", "DT", "SVM", "Stacking"],
    "Accuracy (%)": [93.44, 93.50, 93.92, 90.16, 93.93, 94.92],
    "Positive F-Me (%)": [90.34, 91.16, 87.46, 90.34, 90.85, 92.65],
    "Negative F-Me (%)": [90.34, 89.46, 90.95, 90.34, 90.34, 93.41],
    "Average F-Me (%)": [90.34, 91.44, 80.75, 90.34, 90.14, 92.85],
    "Average Pre (%)": [91.36, 90.76, 79.47, 91.36, 94.12, 94.48],
    "Average Re (%)": [90.12, 88.45, 88.74, 90.12, 91.42, 92.90]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# 1. Accuracy Bar Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Technique', y='Accuracy (%)', data=df, palette='viridis')
plt.title('Accuracy of Different Techniques')
plt.xlabel('Technique')
plt.ylabel('Accuracy (%)')
plt.savefig('accuracy_plot.png')
plt.show()
plt.close()
```

```python
# 2. F-Measure Comparison Plot
plt.figure(figsize=(10, 6))
df_melted = df.melt(id_vars='Technique', value_vars=['Positive F-Me (%)', 'Negative F-Me (%)', 'Average F-Me (%)'],
                    var_name='Class Type', value_name='F-Measure (%)')
sns.barplot(x='Technique', y='F-Measure (%)', hue='Class Type', data=df_melted, palette='magma')
plt.title('F-Measure Comparison')
plt.xlabel('Technique')
plt.ylabel('F-Measure (%)')
plt.legend(title='Class Type')
plt.savefig('fmeasure_comparison_plot.png')
plt.show()
plt.close()
```

```python
# 3. Precision vs Recall Bar Plot
plt.figure(figsize=(10, 6))
df_pr_re = df.melt(id_vars='Technique', value_vars=['Average Pre (%)', 'Average Re (%)'],
                   var_name='Metric', value_name='Value (%)')
sns.barplot(x='Technique', y='Value (%)', hue='Metric', data=df_pr_re, palette='coolwarm')
plt.title('Average Precision vs Recall')
plt.xlabel('Technique')
plt.ylabel('Value (%)')
plt.legend(title='Metric')
plt.savefig('precision_recall_plot.png')
plt.show()
plt.close()
```

```python
# 4. Heatmap of All Metrics
plt.figure(figsize=(10, 8))
heatmap_data = df.set_index('Technique')
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Heatmap of Performance Metrics')
plt.savefig('heatmap_metrics.png')
plt.show()
plt.close()
```

```python
# 5. Line Plot for Accuracy and Average F-Measure
plt.figure(figsize=(10, 6))
plt.plot(df['Technique'], df['Accuracy (%)'], marker='o', label='Accuracy (%)')
plt.plot(df['Technique'], df['Average F-Me (%)'], marker='o', label='Average F-Me (%)')
plt.title('Accuracy vs Average F-Measure')
plt.xlabel('Technique')
plt.ylabel('Percentage (%)')
plt.legend()
plt.grid(True)
plt.savefig('accuracy_vs_fmeasure.png')
plt.show()
plt.close()
```
